# Seaborn:
    
Data Visualization Library

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

NOTE:Seaborn as some built-in dataset

```python
#total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#visualize the data
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
#setting a theme for plot
sns.set_theme()
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
#loading the IRIS dataset
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

#### scatterplot

```python
sns.scatterplot(x='sepal_length',y='petal_length',hue='species',data=iris)
```

```python
sns.scatterplot(x='sepal_width',y='petal_width',hue='species',data=iris)
```

```python
#loading titanic dataset
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
#countplot in dataset
sns.countplot(x='class',data=titanic)
```

```python
sns.countplot(x='survived',data=titanic)
```

#### Bar Chart

```python
sns.barplot(x='sex',y='survived',hue='class',data=titanic)
```

```python
#house price dataset
from sklearn.datasets import load_boston
house_boston = load_boston()
import warnings
warnings.filterwarnings('ignore')
```

```python
house = pd.DataFrame(house_boston.data, columns=house_boston.feature_names)
house['Price'] = house_boston.target
```

```python
house.head()
```

```python
house.shape
```

#### Distribution Plot

```python
sns.distplot(house['Price'])
```

#### correlation 

Postive Correlation

Negative Correlation

#### Heatmap

```python
corr = house.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(corr,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```
